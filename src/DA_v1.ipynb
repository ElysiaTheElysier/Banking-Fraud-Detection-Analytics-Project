```python

df_cards = pd.read_csv(
    r"C:\Ki_4\Domain Banking\Financial Transactions\cards_data.csv",
    dtype={'id': str, 'client_id': str}
)

df_cards['credit_limit'] = df_cards['credit_limit'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)

print(f"Kích thước bảng Cards: {df_cards.shape[0]:,} dòng, {df_cards.shape[1]} cột.")
print("\nThông tin dữ liệu trống (Null):")
print(df_cards.isnull().sum())


df_cards.head()
```

XỬ lý dữ liệu missing data cho Transaction

```python


df['errors'] = df['errors'].fillna('No Error')


df['merchant_state'] = df['merchant_state'].fillna('Online/International')
df['zip'] = df['zip'].fillna('Online/International')

print("Kiểm tra lại xem còn ô nào bị Null không:")
print(df.isnull().sum())
```

Xử lý các file json

```python


with open(r"C:\Ki_4\Domain Banking\Financial Transactions\mcc_codes.json", "r") as f:
    mcc_data = json.load(f)


df_mcc = pd.DataFrame(list(mcc_data.items()), columns=['mcc', 'mcc_description'])

print(f"Tổng số mã ngành hàng: {df_mcc.shape[0]} mã.")


df_mcc.head()
```

```python


with open(r"C:\Ki_4\Domain Banking\Financial Transactions\train_fraud_labels.json", "r") as f:
    fraud_data = json.load(f)


inner_data = list(fraud_data.values())[0]


df_labels = pd.DataFrame(list(inner_data.items()), columns=['id', 'is_fraud'])

print(f"Tổng số nhãn giao dịch: {df_labels.shape[0]:,} dòng.")

print("\nfraud rate")

fraud_distribution = df_labels['is_fraud'].value_counts(normalize=True) * 100
print(fraud_distribution)
```

Tạo bảng fact

```python


df_master = pd.merge(df, df_labels, on='id', how='left')


df_master['is_fraud'] = df_master['is_fraud'].map({'Yes': 1, 'No': 0}).fillna(0).astype(int)

print(f"Kích thước bảng Master: {df_master.shape[0]:,} dòng, {df_master.shape[1]} cột.")
print("\nKiểm tra lại phân phối số lượng Gian lận (1) và Bình thường (0):")
print(df_master['is_fraud'].value_counts())


import os
os.makedirs("clean_data", exist_ok=True)
print("\nĐang lưu file fact_transactions.parquet")
df_master.to_parquet("clean_data/fact_transactions.parquet", index=False)
```

```python
import pandas as pd
from google.cloud import bigquery
from google.oauth2 import service_account



key_path = r"C:\Ki_4\Domain Banking\Financial Transactions\jda-k1-2afad6d8f47e.json"
project_id = "jda-k1" 
dataset_id = "duonglh_bigproject"


credentials = service_account.Credentials.from_service_account_file(key_path)
client = bigquery.Client(credentials=credentials, project=project_id)


tables_to_upload = {
    "dim_mcc": "clean_data/dim_mcc.parquet",
    "dim_users": "clean_data/dim_users.parquet",
    "dim_cards": "clean_data/dim_cards.parquet",
    "fact_transactions": "clean_data/fact_transactions.parquet"
}


for table_name, file_path in tables_to_upload.items():

    df = pd.read_parquet(file_path)
    
    table_ref = f"{project_id}.{dataset_id}.{table_name}"
    

    job_config = bigquery.LoadJobConfig(
        write_disposition=bigquery.WriteDisposition.WRITE_TRUNCATE,
        autodetect=True
    )
    

    job = client.load_table_from_dataframe(df, table_ref, job_config=job_config)
    

    job.result() 


print("\nProcess completed")
```
